## Задача и модель

Рассматривается классическая **множественная линейная регрессионная модель**:

$$
y = X \beta + \varepsilon
$$

где:
- $ y \in \mathbb{R}^n $ — вектор наблюдаемых откликов (цены),
- $ X \in \mathbb{R}^{n \times (k + 1)} $ — фиксированная матрица признаков, содержащая:
  - $ k $ объясняющих количественных переменных,
  - первый столбец из единиц (свободный член),
- $ \beta \in \mathbb{R}^{k + 1} $ — неизвестный вектор коэффициентов модели,
- $ \varepsilon \in \mathbb{R}^n $ — вектор случайных ошибок.


### Предположения модели

Ошибки $ \varepsilon $ удовлетворяют:

- **минимальные**:
  - $ \mathbb{E}[\varepsilon] = 0 $ — несмещённость,
  - $ \text{Var}[\varepsilon] = \sigma^2 \cdot I $ — гомоскедастичность,
  - $ \text{Cov}[\varepsilon_i, \varepsilon_j] = 0 $ при $ i \ne j $ — некоррелированность,

- **нормальные**:
  - $ \varepsilon \sim \mathcal{N}(0, \sigma^2 I) $


## Точечные оценки параметров линейной модели

### Оценка вектора коэффициентов

$$
A = X^{T} X \\
\text{Оценка МНК: } \hat{\beta} = A^{-1} X^{T} y \in \mathbb{R}^{k+1}
$$

### Остатки и остаточная сумма квадратов

$$
\hat{\varepsilon} = y - X \hat{\beta}, \quad
S^2(\hat{\beta}) = \| \hat{\varepsilon} \|^2 = \hat{\varepsilon}^{T} \hat{\varepsilon}
$$

### Несмещённая оценка дисперсии ошибок

Обозначим $ m = k + 1 $. Тогда:

$$
\hat{\sigma}^2 = \frac{S^2(\hat{\beta})}{n - m}
$$


## t-статистика и доверительный интервал для компоненты $\beta_j$

### Проверка гипотезы

Рассматривается точечная гипотеза:

$$
H_0: \beta_j = 0
$$

Используется t-статистика:

$$
T_j = \sqrt{n - m} \cdot \frac{\hat{\beta}_j - \beta_j}{S(\hat{\beta}) \cdot \sqrt{(A^{-1})_{jj}}}
\sim T(n - m)
$$


### Доверительный интервал для $\beta_j$

При уровне доверия $1 - \alpha$:

$$
\hat{\beta}_j \pm \frac{1}{\sqrt{n - m}} \cdot t_{1 - \frac{\alpha}{2}} \cdot S(\hat{\beta}) \cdot \sqrt{(A^{-1})_{jj}}
$$


**Критическая область** зависит от альтернативной гипотезы $ H_1 $:

* **Левосторонняя**: $H_1: \beta_j < 0$

  $$
  T_j < t_{\alpha;\, n - m}
  $$

* **Правосторонняя**: $H_1: \beta_j > 0$

  $$
  T_j > t_{1 - \alpha;\, n - m}
  $$

* **Двусторонняя**: $H_1: \beta_j \ne 0$

  $$
  |T_j| > t_{1 - \frac{\alpha}{2};\, n - m}
  $$


## Доверительный интервал для дисперсии ошибок

При нормальности ошибок, по основной теореме о регрессии:

$$
\frac{(n - m) \cdot \hat{\sigma}^2}{\sigma^2} \sim \chi^2(n - m)
$$

**Интервал уровня** $ 1 - \alpha $:

$$
\left[
\frac{S^2(\hat{\beta})}{\chi^2_{1 - \frac{\alpha}{2}}(n - m)},
\quad
\frac{S^2(\hat{\beta})}{\chi^2_{\frac{\alpha}{2}}(n - m)}
\right]
$$


## F-статистика для проверки гипотез в виде линейных ограничений на $ \beta_j $

Пусть проверяется гипотеза:

$$
H_0: T^T \beta = t_0,
\quad
T \in \mathbb{R}^{m \times r},
\quad
\operatorname{rank} T = r
$$

(т.е. $ r $ независимых линейных условий на параметры).


### F-статистика через разность остатков

Обозначим:
- $ S^2(\hat{\beta}) $ — остаточная сумма квадратов **в полной модели**,
- $ S^2(\hat{\beta}_{t_0, T}) $ — остаточная сумма квадратов **в модели с ограничением** $ T^T \beta = t_0 $

Разложение:

$$
S^2(\hat{\beta}_{t_0, T}) = S^2(\hat{\beta}) +
(\hat{\beta} - \hat{\beta}_{t_0, T})^T A (\hat{\beta} - \hat{\beta}_{t_0, T})
$$

где:

$$
\hat{\beta}_{t_0, T} = \hat{\beta} - A^{-1} T^T D^{-1} (T \hat{\beta} - t_0),
\quad
D = T A^{-1} T^T
$$

Тогда:

$$
F = \frac{n - m}{r} \cdot \frac{S^2(\hat{\beta}_{t_0, T}) - S^2(\hat{\beta})}{S^2(\hat{\beta})}
\sim \mathcal{F}(r,\ n - m)
$$

**Критическая область:** правосторонняя, $ F > f_{1 - \alpha}(r, n - m) $

## Коэффициент корреляции и коэффициент детерминации

### Коэффициент корреляции между $ y $ и $ \hat{y} $

$$
R = \frac{\sum_{i=1}^n (y_i - \bar{y})(\hat{y}_i - \bar{\hat{y}})}
{\sqrt{\sum_{i=1}^n (y_i - \bar{y})^2 \cdot \sum_{i=1}^n (\hat{y}_i - \bar{\hat{y}})^2}}
$$

### Коэффициент детерминации
Это просто $R^2$.

Есть ещё удобная форма записи:
$$R^2 = 1 - \frac{S^2(\hat{\beta})}{\sum (Y_i - \bar{Y})^2}$$

Показывает долю дисперсии $ y $, объяснённую моделью.



```python
import pandas as pd
import numpy as np
from scipy.stats import t, f, chi2
```

```python
df = pd.read_csv("kc_house_data.csv")

alpha = 0.05

X_vars = ['sqft_living', 'sqft_lot']
X_data = df[X_vars].to_numpy()
n = X_data.shape[0]
X = np.column_stack([np.ones(n), X_data])
y = df['price'].to_numpy()

n, m = X.shape  # n — число наблюдений, m — число коэффициентов
print(f"Размерность данных: n = {n}, m = {m}")
print(f"Признаки: β₀ (константа), β₁ (sqft_living), β₂ (sqft_lot)")

A = X.T @ X
print(f"\n rank A = {np.linalg.matrix_rank(A)}")
A_inv = np.linalg.inv(A)
beta_hat = A_inv @ X.T @ y

y_hat = X @ beta_hat
residuals = y - y_hat
S_sq_beta = residuals.T @ residuals

sigma_sq_hat = S_sq_beta / (n - m)

Var_beta_hat = sigma_sq_hat * A_inv
std_errors = np.sqrt(np.diag(Var_beta_hat))

t_crit = t.ppf(1 - alpha/2, df=n - m)
T_vec = beta_hat / std_errors

ci_lower_beta = beta_hat - t_crit * std_errors
ci_upper_beta = beta_hat + t_crit * std_errors

chi2_lower = chi2.ppf(1 - alpha/2, df=n - m)
chi2_upper = chi2.ppf(alpha/2, df=n - m)

ci_lower_sigma_sq = S_sq_beta / chi2_lower
ci_upper_sigma_sq = S_sq_beta / chi2_upper

y_mean = np.mean(y)
TSS = np.sum((y - y_mean) ** 2)
R_squared = 1 - S_sq_beta / TSS

print("\n" + "="*60)
print("ТОЧЕЧНЫЕ ОЦЕНКИ КОЭФФИЦИЕНТОВ (β̂)")
print("="*60)
var_names = ['(константа)', 'sqft_living', 'sqft_lot']
for i, (name, coef) in enumerate(zip(var_names, beta_hat)):
    print(f"β̂_{i} {name}: {coef:.4f}")

print(f"\nТочечная оценка остаточной дисперсии: σ̂² = {sigma_sq_hat:.4f}")
print(f"Стандартная ошибка регрессии: σ̂ = {np.sqrt(sigma_sq_hat):.4f}")

print("\n" + "="*60)
print("ДОВЕРИТЕЛЬНЫЕ ИНТЕРВАЛЫ ДЛЯ КОЭФФИЦИЕНТОВ (уровень 95%)")
print("="*60)
for i, (name, lower, upper) in enumerate(zip(var_names, ci_lower_beta, ci_upper_beta)):
    print(f"β_{i} {name}: ({lower:.4f}, {upper:.4f})")

print("\n" + "="*60)
print("T-СТАТИСТИКИ ДЛЯ ПРОВЕРКИ H₀: βⱼ = 0")
print("="*60)
for i, (name, t_val) in enumerate(zip(var_names, T_vec)):
    p_val = 2 * (1 - t.cdf(np.abs(t_val), df=n - m))
    print(f"T_{i} {name}: {t_val:.4f}, p-value = {p_val:.6f}")

print("\n" + "="*60)
print("ДОВЕРИТЕЛЬНЫЙ ИНТЕРВАЛ ДЛЯ ОСТАТОЧНОЙ ДИСПЕРСИИ (95%)")
print("="*60)
print(f"({ci_lower_sigma_sq:.4f}, {ci_upper_sigma_sq:.4f})")

print("\n" + "="*60)
print("КОЭФФИЦИЕНТ ДЕТЕРМИНАЦИИ")
print("="*60)
print(f"R² = {R_squared:.4f}")
print(f"Модель объясняет {R_squared*100:.2f}% дисперсии цены")
```

```python

print("\nГипотеза 1: Чем больше жилая площадь, тем больше цена")
print("   H₀: β₁ = 0 vs H₁: β₁ > 0 (sqft_living)")

T_sqft_living = T_vec[1]
p_sqft_living = 1 - t.cdf(T_sqft_living, df=n - m)
t_crit_right = t.ppf(1 - alpha, df=n - m)
decision1 = "Отклоняем H₀ ✓" if T_sqft_living > t_crit_right else "Нет оснований отклонить H₀"

print(f"   T = {T_sqft_living:.4f}")
print(f"   t_крит (α={alpha}) = {t_crit_right:.4f}")
print(f"   p-value = {p_sqft_living:.6f}")

print(f"   ⇒ {decision1}")
if T_sqft_living > t_crit_right:
    print("   Вывод: С увеличением жилой площади цена статистически значимо растёт")

print("\nГипотеза 2: Цена зависит от площади участка")
print("   H₀: β₂ = 0 vs H₁: β₂ ≠ 0 (sqft_lot)")

T_sqft_lot = T_vec[2]
p_sqft_lot = 2 * (1 - t.cdf(np.abs(T_sqft_lot), df=n - m))
t_crit_two = t.ppf(1 - alpha/2, df=n - m)
decision2 = "Отклоняем H₀ ✓" if np.abs(T_sqft_lot) > t_crit_two else "Нет оснований отклонить H₀"

print(f"   T = {T_sqft_lot:.4f}")
print(f"   |t_крит| (α={alpha}) = {t_crit_two:.4f}")
print(f"   p-value = {p_sqft_lot:.6f}")
print(f"   ⇒ {decision2}")
if np.abs(T_sqft_lot) > t_crit_two:
    print("   Вывод: Цена статистически значимо зависит от площади участка")

print("\nГипотеза 3: Совместная значимость sqft_living и sqft_lot")
print("   H₀: β₁ = β₂ = 0 vs H₁: не все βⱼ = 0")

T_matrix = np.array([
    [0, 1, 0],
    [0, 0, 1]
]).T
t0_vec = np.zeros(2)
r = T_matrix.shape[1]

D = T_matrix.T @ A_inv @ T_matrix

beta_restricted = beta_hat - A_inv @ T_matrix @ np.linalg.inv(D) @ (T_matrix.T @ beta_hat - t0_vec)

delta = beta_hat - beta_restricted
delta_SSq = delta.T @ A @ delta
S_sq_restricted = S_sq_beta + delta_SSq

F_stat = (n - m) / r * (S_sq_restricted - S_sq_beta) / S_sq_beta
F_crit = f.ppf(1 - alpha, dfn=r, dfd=n - m)
p_f = 1 - f.cdf(F_stat, dfn=r, dfd=n - m)
decision3 = "Отклоняем H₀ ✓" if F_stat > F_crit else "Нет оснований отклонить H₀"

print(f"   F = {F_stat:.4f}")
print(f"   F_крит({r}, {n-m}) при α={alpha} = {F_crit:.4f}")
print(f"   p-value = {p_f:.10f}")
print(f"   ⇒ {decision3}")
if F_stat > F_crit:
    print("   Вывод: Хотя бы один из коэффициентов (sqft_living, sqft_lot) статистически отличен от нуля")
```

## Формулировка задачи

**Цель:**
Проверить, одинаковы ли средние значения суммарных баллов среди различных **этнических групп**.


## Теоретическая модель: однофакторный ANOVA

### Обозначения:

* Пусть есть $k$ групп, каждая с размером $n_j$
* $y_{ij}$ — $i$-е наблюдение в $j$-й группе
* $y_{*j}$ — среднее в $j$-й группе
* $y_{**}$ — общее среднее

### Гипотезы

* **Нулевая гипотеза**:

  $$
  H_0: \mu_1 = \mu_2 = \cdots = \mu_k
  $$

* **Альтернатива**:

  $$
  H_1: \exists\, i \ne j: \mu_i \ne \mu_j
  $$


### Статистика ANOVA

Обозначим:

* **SST (Total Sum of Squares)**:

  $$
  \text{SST} = \sum_{j=1}^k \sum_{i=1}^{n_j} (y_{ij} - y_{**})^2
  $$

* **SSB (Between-Groups Sum of Squares)**:

  $$
  \text{SSB} = \sum_{j=1}^k n_j (y_{*j} - y_{**})^2
  $$

* **SSW (Within-Groups Sum of Squares)**:

  $$
  \text{SSW} = \sum_{j=1}^k \sum_{i=1}^{n_j} (y_{ij} - y_{*j})^2
  $$


### F-статистика

$$
F = \frac{\text{SSB} / (k - 1)}{\text{SSW} / (n - k)} \sim \mathcal{F}(k - 1, n - k)
$$


### Правило принятия решения:

* Если $F > F_{1 - \alpha}(k - 1,\ n - k)$, **отвергаем $H_0$**
* Иначе — **нет оснований отвергнуть $H_0$**

```python
df = pd.read_csv("exams_dataset.csv")
df['score'] = df[['math score', 'reading score', 'writing score']].sum(axis=1)

groups = df['race/ethnicity'].unique()
k = len(groups)

group_data = [df[df['race/ethnicity'] == g]['score'].values for g in groups]
n_j = [len(g) for g in group_data]
n = sum(n_j)

print(f"Фактор: race/ethnicity")
print(f"Число групп: k = {k}")
print(f"Общее число наблюдений: n = {n}")
print(f"Размеры групп: {dict(zip(groups, n_j))}")

means_j = [np.mean(g) for g in group_data]  # ȳ_j — среднее в j-й группе
mean_total = np.mean(df['score'])           # ȳ — общее среднее


SSB = sum(nj * (mj - mean_total)**2 for nj, mj in zip(n_j, means_j))

SSW = sum(np.sum((g - mj)**2) for g, mj in zip(group_data, means_j))

SST = SSB + SSW

dfB = k - 1       # степени свободы между группами
dfW = n - k       # степени свободы внутри групп

MSB = SSB / dfB  
MSW = SSW / dfW 

F_stat = MSB / MSW
p_value = 1 - f.cdf(F_stat, dfB, dfW)

alpha = 0.05
F_crit = f.ppf(1 - alpha, dfB, dfW)

print("\n" + "="*60)
print("ОДНОФАКТОРНЫЙ ДИСПЕРСИОННЫЙ АНАЛИЗ")
print("="*60)

print("\nТаблица ANOVA:")
print("-"*70)
print(f"{'Источник':<20}{'SS':>12}{'df':>8}{'MS':>15}{'F':>12}")
print("-"*70)
print(f"{'Между группами':<20}{SSB:>12.2f}{dfB:>8}{MSB:>15.2f}{F_stat:>12.4f}")
print(f"{'Внутри групп':<20}{SSW:>12.2f}{dfW:>8}{MSW:>15.2f}{'-':>12}")
print(f"{'Всего':<20}{SST:>12.2f}{n-1:>8}{'-':>15}{'-':>12}")
print("-"*70)

print(f"\nРезультаты теста:")
print(f"   F-статистика: {F_stat:.4f}")
print(f"   F_крит({dfB}, {dfW}) при α = {alpha}: {F_crit:.4f}")
print(f"   p-value: {p_value:.6f}")


if F_stat > F_crit:
    print(f"\nРешение: F = {F_stat:.4f} > F_крит = {F_crit:.4f}")
    print("   ⇒ H₀ отвергается: различия между группами статистически значимы")
    print("   Средние суммарные баллы статистически различаются между этническими группами")
else:
    print(f"\nРешение: F = {F_stat:.4f} ≤ F_крит = {F_crit:.4f}")
    print("   ⇒ Нет оснований отвергнуть H₀: различия незначимы")
```

```python
print("\n" + "="*60)
print("СРЕДНИЕ ЗНАЧЕНИЯ ПО ГРУППАМ")
print("="*60)
print(f"{'Группа':<15}{'n_j':>8}{'Среднее':>12}{'Std':>12}")
print("-"*50)
for g, data, nj, mj in zip(groups, group_data, n_j, means_j):
    std_j = np.std(data, ddof=1)
    print(f"{g:<15}{nj:>8}{mj:>12.2f}{std_j:>12.2f}")
print("-"*50)
print(f"{'Общее':<15}{n:>8}{mean_total:>12.2f}")

print(f"\n✓ Проверка: SST = SSB + SSW")
print(f"  {SST:.2f} = {SSB:.2f} + {SSW:.2f} = {SSB + SSW:.2f}")
```
